```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
guess = 0.0
cube = 27
increment = 0.1
epsilon = 0.1 # Error Tolerance;
```

```python
# Finding the approximate value
cnt=30
list1=[]
list2=[]
for i in range(0,cnt):
    if(abs (guess**3 - cube)) >= epsilon:
        guess+=increment
        list1.append(guess)
        list2.append(guess**3 - cube)

print(list1)
print(list2)
```

```python
plt.plot(list1)
plt.plot(list2)
plt.xlabel('guess')
plt.ylabel('Error')
plt.show()
```
